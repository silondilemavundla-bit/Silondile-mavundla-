```python
import sqlite3
import pandas as pd

# Create in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    city TEXT,
    income REAL
)
""")

cursor.execute("""
CREATE TABLE transactions (
    transaction_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    transaction_date TEXT,
    amount REAL,
    transaction_type TEXT,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
)
""")
```

Time-series transaction trends

```python
customers_data = [
    (1, "Alice", 29, "Johannesburg", 45000),
    (2, "Bob", 45, "Cape Town", 72000),
    (3, "Charlie", 34, "Durban", 38000),
    (4, "David", 52, "Pretoria", 95000),
    (5, "Emma", 40, "Johannesburg", 60000)
]

cursor.executemany("INSERT INTO customers VALUES (?, ?, ?, ?, ?)", customers_data)

transactions_data = [
    (1, 1, "2024-01-10", 5000, "purchase"),
    (2, 1, "2024-01-12", 15000, "transfer"),
    (3, 2, "2024-01-11", 25000, "purchase"),
    (4, 3, "2024-01-09", 7000, "withdrawal"),
    (5, 4, "2024-01-15", 45000, "transfer"),
    (6, 5, "2024-01-16", 12000, "purchase"),
    (7, 1, "2024-01-18", 22000, "transfer"),
    (8, 3, "2024-01-20", 3000, "purchase")
]

cursor.executemany("INSERT INTO transactions VALUES (?, ?, ?, ?, ?)", transactions_data)

conn.commit()
```

Fraud detection

```python
query = """
SELECT c.name,
       SUM(t.amount) AS total_spent
FROM customers c
JOIN transactions t ON c.customer_id = t.customer_id
GROUP BY c.name
ORDER BY total_spent DESC
"""
pd.read_sql(query, conn)
```

Rolling 7-day averages

```python
query = """
SELECT c.name,
       COUNT(*) AS large_transfers
FROM customers c
JOIN transactions t ON c.customer_id = t.customer_id
WHERE t.transaction_type = 'transfer'
  AND t.amount > 20000
GROUP BY c.name
"""
pd.read_sql(query, conn)
```

Cohort analysis

```python
query = """
SELECT name,
       total_spent,
       RANK() OVER (ORDER BY total_spent DESC) AS spending_rank
FROM (
    SELECT c.name,
           SUM(t.amount) AS total_spent
    FROM customers c
    JOIN transactions t ON c.customer_id = t.customer_id
    GROUP BY c.name
)
"""
pd.read_sql(query, conn)
```

Customer lifetime value calculation

```python
query = """
SELECT c.name,
       SUM(t.amount) AS total_spent,
       SUM(CASE WHEN t.amount > 20000 THEN 1 ELSE 0 END) AS large_tx_count,
       CASE
           WHEN SUM(t.amount) > 50000 THEN 'High Value'
           WHEN SUM(t.amount) > 20000 THEN 'Medium Value'
           ELSE 'Low Value'
       END AS customer_segment
FROM customers c
JOIN transactions t ON c.customer_id = t.customer_id
GROUP BY c.name
"""
pd.read_sql(query, conn)
```
